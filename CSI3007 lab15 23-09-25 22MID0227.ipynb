**New histogram being plotted for the greyscale image and difference is observed between the original image and the greyscale image as more darker datapoints are present in the greyscale image**

```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np

color_image = Image.open('/home/matlab/Downloads/bug.jpg')

grayscale_image = color_image.convert('L')

grayscale_image_array = np.asarray(grayscale_image)

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)  
plt.imshow(grayscale_image_array, cmap='gray')
plt.title("Grayscale Image")
plt.axis('off')


plt.subplot(1, 2, 2) 
plt.hist(grayscale_image_array.ravel(), bins=range(256), fc='k', ec='k')
plt.title('Pixel Intensity Histogram')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')

plt.show()
```
